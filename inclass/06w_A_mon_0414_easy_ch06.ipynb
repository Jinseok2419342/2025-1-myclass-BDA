## mpg 데이터 활용 groupby() agg()

```python
mpg = pd.read_csv('data/mpg.csv')
mpg.info()
```

```python
mpg.columns
```

```python
mpg.sample(frac=.01)
```

```python
mpg.groupby(['manufacturer', 'drv']).agg(mean_cty = ('cty', 'mean'))
```

```python
mpg.groupby(['drv', 'manufacturer']).agg(mean_cty = ('cty', 'mean'))
```

```python
mpg.groupby(['manufacturer', 'category']).agg(mean_cty = ('cty', 'mean'), mean_hwy = ('hwy', 'mean'))
```

```python
mpg.groupby(['category', 'manufacturer']).agg(mean_cty = ('cty', 'mean'), mean_hwy = ('hwy', 'mean'))
```

```python
mpg.query('manufacturer == "audi"').groupby(['drv']).agg(n = ('drv', 'count'))
```

```python
mpg.query('manufacturer == "chevrolet"').groupby(['drv']).agg(n = ('drv', 'count'))
```

```python
mpg.query('manufacturer == "chevrolet"') \
   .groupby(['drv']) \
   .agg(n = ('drv', 'count'))
```

```python
mpg.query('manufacturer == "chevrolet"').groupby(['drv']).count()
```

```python
mpg.query('manufacturer == "chevrolet"').groupby(['drv']).count()[["model"]]
```

```python
mpg.query('manufacturer == "chevrolet"').groupby(['drv']) \
   .count()[["model"]].rename(columns={'model': 'n'})
```

```python
mpg.groupby('drv').agg(n = ('drv', 'count'))
```

```python
mpg[['drv', 'cyl']].value_counts()
```

```python
mpg[['drv', 'cyl']].value_counts().to_frame()
```

```python
mpg.groupby(['drv', 'cyl']).agg('count')[['model']]
```

```python
mpg[['drv', 'cyl']].value_counts().to_frame()
```

```python
mpg[['drv', 'cyl']].value_counts().to_frame().query('count > 40')
```

```python
mpg['drv'].value_counts().to_frame()
```

```python
mpg['drv'].value_counts().to_frame('n')
```

```python
mpg['drv'].value_counts().to_frame('n').rename_axis('drive')
```

```python
mpg.query('category == "suv"') \
   .assign(total = (mpg.hwy + mpg.cty) / 2) \
   .groupby('manufacturer') \
   .agg(mean_total = ('total', 'mean')) \
   .sort_values('mean_total', ascending=False) \
   .head(5)
```

```python
mpg.query('category == "suv"') \
   .assign(total = (mpg.hwy + mpg.cty) / 2) \
   .groupby('manufacturer') \
   .agg(mean_total = ('total', 'mean')) \
   .sort_values('mean_total', ascending=False)
```

## 데이터 합치기

```python
# 중간고사 데이터 만들기
test1 = pd.DataFrame({'id'      : [1, 2, 3, 4, 5],
                      'midterm' : [60, 80, 70, 90, 85]})

# 기말고사 데이터 만들기
test2 = pd.DataFrame({'id'    : [1, 2, 3, 4, 5],
                      'final' : [70, 83, 65, 95, 80]})
```

```python
test1
```

```python
test2
```

```python
pd.merge(test1, test2)
```

```python
pd.merge(test1, test2, on = 'id')
```

```python
test2.merge(test1, on = 'id', how='inner')
```
